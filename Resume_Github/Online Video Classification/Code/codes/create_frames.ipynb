```
def split_video_filename(video_path):
    """Given a full path to a video, return its parts."""
    
    parts = video_path.split('/')
    filename = parts[6]
    filename_no_ext = filename.split('.')[0]
    classname = parts[5]
    train_or_test = parts[4]
    
    #print('parts : ',parts)
    
    return train_or_test, classname, filename_no_ext, filename
```

```
def check_if_extracted(video_parts):
    
    train_or_test, classname, filename_no_ext, _ = video_parts
    
    temp_str = 'gdrive/My Drive/ADBI_Capstone_Video_Classification/data/'+train_or_test+'/'+ classname+'/'+ filename_no_ext + '-0001.jpg'
    
    return bool(os.path.exists(temp_str))
```

```
def get_total_number_of_frames(video_parts):
    
    train_or_test, classname, filename_no_ext, _ = video_parts
    
    vid_file = 'gdrive/My Drive/ADBI_Capstone_Video_Classification/data/'+train_or_test +'/'+classname+'/'+ filename_no_ext + '*.jpg'
    generated_files = glob.glob(vid_file)
    
    return len(generated_files)
```

```
def convert_videos_to_images():
    
    print('Inside function convert_videos_to_images()')
    
    data_file = []
    folders = ['train_videos', 'test_videos']

    for folder in folders:
        
        class_folders = glob.glob('gdrive/My Drive/ADBI_Capstone_Video_Classification/data/'+folder+'/*')
        
        print('class_folder : ',class_folders)
        
        for vid_class in class_folders:
            
            class_files = glob.glob(vid_class+'/*.avi')

            for video_path in class_files:
              
                video_parts = split_video_filename(video_path)
                
                train_or_test, classname, filename_no_ext, filename = video_parts

                if not check_if_extracted(video_parts):
                    
                    src = 'gdrive/My\ Drive/ADBI_Capstone_Video_Classification/data/'+train_or_test +'/'+ classname+'/'+ filename
                    
                    if train_or_test == 'train_videos':
                      dest_train_or_test = 'train'
                    else:
                      dest_train_or_test = 'test'
                    
                    temp_dir = 'gdrive/My\ Drive/ADBI_Capstone_Video_Classification/data/'+dest_train_or_test+'/'+ classname+'/'
                    !mkdir -p {temp_dir}
                    
                    dest = 'gdrive/My\ Drive/ADBI_Capstone_Video_Classification/data/'+dest_train_or_test+'/'+ classname+'/'+ filename_no_ext + '-%04d.jpg'
                    
                    !ffmpeg -i {src} {dest}
                
                nb_frames = get_total_number_of_frames(video_parts)

                data_file.append([dest_train_or_test, classname, filename_no_ext, nb_frames])

                print("\n\n Generated %d frames for %s" % (nb_frames, filename_no_ext))

    with open('gdrive/My Drive/ADBI_Capstone_Video_Classification/temp_data_file.csv', 'w') as fout:
        writer = csv.writer(fout)
        writer.writerows(data_file)

    print("Extracted and wrote %d video files." % (len(data_file)))
    
```

```
if __name__ == '__main__':
    convert_videos_to_images()
```
